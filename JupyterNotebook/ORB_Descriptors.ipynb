```python
#USING Brute-Force Matching with ORB Descriptors 
import yaml
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Path Config
path = "dataset-1.0/"
subpath_annotations = "annotations/"
subpath_images = "images/"
```

```python
image_file="004_image.png"
img = cv2.imread(path+subpath_images+image_file)
```

```python
plt.imshow(img)
plt.show()
```

```python
# Get all features
boundary = ([15, 90, 20], [110, 230, 80])
lower = np.array(boundary[0],dtype=np.uint8)
upper = np.array(boundary[1],dtype=np.uint8)
```

```python
annotation_image_file="004_annotation.png"
imgannotation = cv2.imread(path+subpath_annotations+annotation_image_file)
```

```python
plt.imshow(imgannotation)
plt.show()
```

```python
annotations = path+subpath_annotations+"004_annotation.yaml"
annotations_json = []
annotations_json.append(yaml.load(open(annotations, 'r')))
print(annotations_json)
```

```python
annotations = annotations_json
all_descriptors = np.arange(32).reshape(1,32)
for annotation in annotations:
    regions = annotation['annotation']
    for region in regions:
        points = region['points']
        px = points['x']
        py = points['y']
        if(type(px) is list):
            pts = np.array([[int(x),int(y)] for x,y in zip(px,py)], dtype=np.int32)
            mask_region = np.zeros(img.shape, dtype=np.uint8)
            cv2.fillConvexPoly(mask_region, pts, (255,255,255))
            masked_region = cv2.bitwise_and(img,mask_region)
            mask_color = cv2.inRange(masked_region, lower, upper)
            masked_color = cv2.bitwise_and(masked_region, masked_region, mask = mask_color)
            plt.imshow(masked_color)
            plt.show()

            
            #ORB DESCRIPTORS 32
            orb = cv2.ORB_create()
            kps, descriptors = orb.detectAndCompute(masked_color,None)
            
            print("SIFT KEYPOINTS")
            #SHOW SIFT KEYPOINTS
            cv2.drawKeypoints(masked_color,kps,masked_color)
            plt.imshow(masked_color)
            plt.show()
            all_descriptors = np.append(all_descriptors, descriptors, axis=0)
            print(all_descriptors)
            print(all_descriptors.shape)
print("========= ALL IMAGE DESCRIPTORS =========")
print(all_descriptors)
print(all_descriptors.shape)
            
```
